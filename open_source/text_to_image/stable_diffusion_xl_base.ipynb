```python
from diffusers import DiffusionPipeline
import torch

# pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float32, use_safetensors=True)
```

```python
prompt = "An astronaut riding a green horse"

images = pipe(prompt=prompt).images[0]
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# img = mpimg.imread("astronaut_rides_horse.png")
plt.imshow(images)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()

import matplotlib.pyplot as plt


# Save the displayed image to a file (e.g., "output_image.png")
# plt.savefig("img/An astronaut riding a green horse_2.png", bbox_inches='tight', pad_inches=0)
plt.savefig("img/An astronaut riding a green horse_2.png")

# Show the image
plt.show()

  
```
