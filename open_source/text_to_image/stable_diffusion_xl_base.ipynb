```python
from diffusers import DiffusionPipeline
import torch

# pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
# Tar ca 30 min å laste inn modellen
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float32, use_safetensors=True)
```

```python
prompt = "A football fan at Old Trafford, eating a hamburger in one hand and holding a beer in the other, is celebrating Ole Gunnar Solskjær's best goal."
prompt = "Create a high-resolution and realistic image of a nurse standing heroically. The nurse should be in a traditional nursing uniform, complete with a white tunic and a nurse's cap. On the back, the nurse should have a bright red Superman cape fluttering in the wind. The background should be a hospital environment, preferably with visible medical equipment and a white corridor. The image should be in daylight with a sense of optimism and strength. Emphasize details like textures on the clothes, realistic lighting effects, and an expression of care and confidence in the nurse's facial features. Focus on creating an image that honors the nursing profession with a creative twist."

# Tar 22 / 30 min tidligere
images = pipe(prompt=prompt).images
# images = pipe(prompt=prompt).images[0]
# 
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Display the image
plt.imshow(images[0])
plt.axis('off')  # Turn off axis labels and ticks
plt.show()

# Display the image
# Save the displayed image to a file (e.g., "output_image.png")
plt.savefig("img/A Man Utd fan.png", bbox_inches='tight', pad_inches=0)

# Show the image
plt.show()

  
```
